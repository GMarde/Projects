# Manipulação de Strings
________________________________________

```python
igm.info()
```

Observe que algumas colunas, como por ex 'área', estão como type "object" mesmo contendo números

```python
igm['area'].sample(5)
```

```python
# Esse código não funciona porque a vírgula separa os milhares
#igm['area'].astype(float)
```

```python
igm['area'].str.replace(',','').astype(float).sample(10)
```

```python
igm['area'] = igm['area'].str.replace(',','').astype(float)
```

```python
igm['populacao'].sample(10)
```

```python
igm['populacao'] = igm['populacao'].str.replace(',','')
```

```python
# Essa linha não funciona porque há valores entre parenteses
#igm['populacao'] = igm['populacao'].astype(float)
```

```python
igm['populacao'] = igm['populacao'].str.replace('.','')
```

```python
valor_problema = '41.487(1)'
```

```python
valor_problema.split('(')
```

```python
# O código valor_problema.split('(') gera uma lista, então pegamos apenas o primeiro valor da lista
valor_problema.split('(')[0]
```

```python
igm['populacao'] = igm['populacao'].str.split('(')
```

```python
igm['populacao'].sample(10)
```

```python
igm['populacao'] = igm['populacao'].str[0]
```

```python
igm['populacao'].sample(10)
```

```python
igm['populacao'] = igm['populacao'].str[0]
```

```python
igm['populacao'] = igm['populacao'].astype(float)
```
